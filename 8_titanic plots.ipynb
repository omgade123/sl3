```python
male_passengers = titanic[titanic['sex'] == 'male']
female_passengers = titanic[titanic['sex'] == 'female']

# Select numerical columns for analysis
numerical_columns = ['age', 'fare', 'pclass']

# Calculate correlation matrices for male and female passengers  
#A correlation matrix is a table that displays the correlation coefficients between multiple variables. 
corr_male = male_passengers[numerical_columns].corr()
corr_female = female_passengers[numerical_columns].corr()

# Plot heatmaps
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sn.heatmap(corr_male, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap - Male Passengers')

plt.subplot(1, 2, 2)
sn.heatmap(corr_female, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap - Female Passengers')

plt.tight_layout()
plt.show()
```
