```python
import numpy as numpy
import pandas as pd
```

```python
df = pd.read_csv("D:\SEM 6\Software Lab 3\Practical Dataset\Practical 4 Boston Housing\HousingData.csv")
df
```

```python
df.isnull().sum()
```

```python
df.interpolate(method = 'linear',limit_direction = 'forward', inplace = True)
X = df
Y = df
X = X.drop(['MEDV'], axis = 1)
X # Feature Values
```

```python
Y = df['MEDV'] # Target Values
Y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size = 0.2,random_state = 40)
```

```python
X_train
```

```python
X_test
```

```python
Y_test
```

```python
Y_train
```

```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm.fit(X_train,Y_train)
score = lm.score(X_test,Y_test)  #model Evaluation, This return the value of R^2 coefficient 
print(score)
```

```python
Y_train_pred = lm.predict(X_train)
Y_test_pred = lm.predict(X_test)
df1 = pd.DataFrame(Y_train)
from sklearn.metrics import mean_squared_error as mse
error = mse(Y_test_pred,Y_test)
print(error)
mse = mse(Y_train_pred,Y_train)
print("mse is ", mse)
rmae = mse**0.5
print("rmae is ", rmae)
```

```python
import matplotlib.pyplot as plt
plt.scatter(Y_train_pred, Y_train, c = 'blue', marker = 'o', label = 'Training Data')
plt.scatter(Y_test_pred, Y_test, c = 'green', marker = 's', label = 'Test Data')
plt.xlabel("Predicted")
plt.ylabel("Actual Value")
plt.plot()
plt.legend()
plt.show()
```
